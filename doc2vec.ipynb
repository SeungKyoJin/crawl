토크나이징을 위한 KoNLpy에서 twitter 모듈 함수화

```python
twitter = Twitter()
def tokenizer_twitter_morphs(doc):
    return twitter.morphs(doc)

def tokenizer_twitter_noun(doc):
    return twitter.nouns(doc)

def tokenizer_twitter_pos(doc):
    return twitter.pos(doc, norm=True, stem=True)

komoran = Komoran()
def tokenizer_noun(doc):
    return komoran.nouns(doc)

def tokenizer_morphs(doc):
    return komoran.porphs(doc)
```

토큰화 한 것들을 df_news에 추가

```python
df_news['token_title'] = df_news['title'].apply(tokenizer_twitter_morphs)
df_news['token_contents'] = df_news['contents'].apply(tokenizer_twitter_morphs)
df_news['token_keywords'] = df_news['keywords'].apply(tokenizer_twitter_morphs)
```

```python
# 명사만 추출
df_news['token_title_1'] = df_news['title'].apply(tokenizer_twitter_noun)
df_news['token_contents_1'] = df_news['contents'].apply(tokenizer_twitter_noun)
df_news['token_keywords_1'] = df_news['keywords'].apply(tokenizer_twitter_noun)
```

```python
df_news.head()
```

토큰화 한 것들 리스트화

```python
tokens_title = [t for d in df_news['token_title_1'] for t in d]
tokens_contents = [t for d in df_news['token_contents_1'] for t in d]
tokens_keywords = [t for d in df_news['token_keywords_1'] for t in d]
```

doc2vec 모델을 학습하기 위한 TaggedDocument 형태로 전환

```python
tagged_train_docs = [TaggedDocument(d, c) for d, c in df_train[['token_contents', 'token_keywords']].values]
tagged_test_docs = [TaggedDocument(d, c) for d, c in df_test[['token_contents', 'token_keywords']].values]
```

```python
print(len(tagged_train_docs), len(tagged_test_docs))
```

```python
import multiprocessing
cores = multiprocessing.cpu_count()
cores
```

```python
import logging
logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO)
```

```python
from gensim.models.doc2vec import Doc2Vec, TaggedDocument
from time import time
```

doc2vec 모델의 하이퍼파라미터 선언

```python
doc_vectorizer = Doc2Vec(
    dm=0,            # PV-DBOW / default 1
    dbow_words=1,    # w2v simultaneous with DBOW d2v / default 0
    window=8,        # distance between the predicted word and context words
    size=300,        # vector size
    alpha=0.025,     # learning-rate
    seed=1234,
    min_count=20,    # ignore with freq lower
    min_alpha=0.025, # min learning-rate
    workers=cores,   # multi cpu
    hs = 1,          # hierarchical softmax / default 0
    negative = 10,   # negative sampling / default 5
)
```

```python
doc_vectorizer.build_vocab(tagged_train_docs)
print(str(doc_vectorizer))
```

```python
doc_vectorizer.corpus_count, doc_vectorizer.iter
```

학습시작!

```python
start = time()

for epoch in range(10):
    doc_vectorizer.train(tagged_train_docs, total_examples=doc_vectorizer.corpus_count, epochs=doc_vectorizer.iter)
    doc_vectorizer.alpha -= 0.002
    doc_vectorizer.min_alpha = doc_vectorizer.alpha

end = time()

print("Time: {}".format(end-start))
```

모델 저장

```python
model_name_1 = 'Doc2Vec(dbow+w,d300,n10,hs,w8,mc20,s0.001,t8)'
doc_vectorizer.save(model_name_1)
```
